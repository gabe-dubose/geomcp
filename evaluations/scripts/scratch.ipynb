```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load data
high_init_low_oA_df = pd.read_csv('../data/evolving_pleio_high_init_low_oA.csv')
high_init_high_oA_df = pd.read_csv('../data/evolving_pleio_high_init_high_oA.csv')
```

```python
sns.set_style('whitegrid')
fig, [ax1, ax2] = plt.subplots(1,2, figsize=(10, 5), sharex=True, sharey=True)

# low oA, high initial pleiotropy
for index, row in high_init_low_oA_df.iterrows():
    sns.lineplot(list(row), ax=ax1, legend=False, color = 'tab:gray', alpha=0.5)
# plot mean dynamics
sns.lineplot(list(high_init_low_oA_df.mean(axis=0)), ax=ax1, legend=False, color='black')

ax1.tick_params(axis='both', labelsize=12)
ax1.set_ylabel('Mean r', fontsize=12)
ax1.set_xlabel('Generations', fontsize=12)

# high oA, high initial pleiotropy
for index, row in high_init_high_oA_df.iterrows():
    sns.lineplot(list(row), ax=ax2, legend=False, color = 'tab:gray', alpha=0.5)
# plot mean dynamics
sns.lineplot(list(high_init_high_oA_df.mean(axis=0)), ax=ax2, legend=False, color='black')

ax1.set_xlim(0, 100)
ax2.set_xlim(0, 100)
```
